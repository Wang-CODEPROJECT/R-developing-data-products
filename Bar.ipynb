```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
path = 'Canada.xlsx'
df_can = pd.read_excel(path, sheet_name='Canada by Citizenship', skiprows=20, skipfooter=2)
df_can.head()
```

```python
df_can.drop(['Type', 'Coverage', 'AREA', 'REG', 'DEV'], axis=1, inplace=True)
df_can.head(1)
```

```python
df_can.rename(columns={'OdName':'Country', 'AreaName':'Continet', 'RegName':'Region'},inplace=True)
df_can.head(1)
```

```python
df_can['Total'] = df_can.sum(axis=1)
```

```python
df_can.columns = df_can.columns.map(str)
years = list(map(str, range(1980, 2014)))
```

```python
df_can.set_index('Country', inplace=True)
df_can.index.name = None
df_can.head()
```

```python
df_can.sort_values('Total', ascending=False, inplace=True)
```

```python
df_can.head()
```

```python
df_Ice = df_can.loc['Iceland', years]
df_Ice.head()
```

```python
df_Ice.plot(kind='bar', figsize=(10, 6), rot=60)

plt.annotate('1111',
            xy=(32, 70),
            xytext=(28, 20),
            xycoords='data',
            arrowprops=dict(arrowstyle='->', connectionstyle='arc3', color='blue', lw=2))

plt.annotate('2008-2011',
            xy=(28, 30),
            rotation=60,
            va='bottom',
            ha='left')
```
